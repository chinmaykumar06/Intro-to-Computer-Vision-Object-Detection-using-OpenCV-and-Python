### How to Tune Cascade Classifier

(https://docs.opencv.org/2.4/modules/objdetect/doc/cascade_classification.html#cascadeclassifier-detectmultiscale)

<b>detectMultiscale</b>(input image, <b>Scale Factor</b>, <b>Min Neighbors</b>)
- <b>Scale Factor</b> Parameter specifying how much the image size is reduced at each image scale.


- <b>Min Neighbors</b> Parameter specifying how many neighbors each candidate rectangle should have to retain it.

### Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Create Classifier

```python
face_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
```

### Open Image

```python
image = cv2.imread('Images/eye_face.jpg')
fix_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

### It's a kind of Magic

```python
def detect_face(fix_img):
    face_rects = face_classifier.detectMultiScale(fix_img)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(fix_img,
                     (x,y),
                     (x+w, y+h),
                     (255,0,0),
                     10)
    return fix_img
```

```python
result = detect_face(fix_img)
plt.imshow(result)
plt.savefig('faces.png', bbox_inches='tight')
```
